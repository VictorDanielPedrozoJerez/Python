# Diferenciación Numérica

## **Problema 1**
 Dada la tabla de valores para una función $f(x)$:

| $x$ | $f(x)$ |
|-----|--------|
| 0.5 | 0.4794 |
| 0.6 | 0.5646 |
| 0.7 | 0.6442 |

Utilice Python para aproximar $f'(0.5)$ usando una fórmula de diferencias hacia adelante de dos puntos, $f'(0.7)$ usando una fórmula de diferencias hacia atrás de dos puntos, y $f'(0.6)$ usando una fórmula de diferencias centrales de dos puntos. Reporte sus valores y el código. Comente sobre la precisión esperada si $h=0.1$.







* Fórmula de Diferencias Hacia Adelante:

$$
f'(x) \approx \frac{f(x+h) - f(x)}{h}
$$
*  Fórmula de Diferencias Hacia Atrás

$$
f'(x) \approx \frac{f(x) - f(x-h)}{h}
$$

* Fórmula de Diferencias Centrales

$$
f'(x) \approx \frac{f(x+h) - f(x-h)}{2h}
$$

## Desarrollo:

```python
f = [0.4794, 0.5646, 0.6442]
h = 0.1

# Diferencia hacia adelante (f'(0.5))
adelante = (f[1] - f[0]) / h
# Diferencia hacia atrás (f'(0.7))
atras = (f[2] - f[1]) / h
# Diferencia central (f'(0.6))
central = (f[2] - f[0]) / (2 * h)

print(" f'(0.5) (hacia adelante): ", adelante)
print(" f'(0.7) (hacia atrás): ", atras)
print(" f'(0.6) (central): ", central)
```

La precisión de las fórmulas de diferencias finitas depende del tamaño del paso h y del método empleado. Las diferencias hacia adelante/atrás  ofrecen baja precisión, con errores proporcionales a 0.1, mientras que las diferencias centrales son más precisas, reduciendo el error a términos de 0.01. Sin embargo, al disminuir h excesivamente, domina el error de redondeo por pérdida de cifras significativas.

## **Problema 2**

Para $f(x) = e^x - 2x^2 + 3x - 1$, con datos $x=[0.0, 0.2, 0.4]$ y $f(x)=[0.00000, 0.74140, 1.3718]$.

1. Aproxime $f'(0.0)$, $f'(0.2)$, $f'(0.4)$ usando las fórmulas de tres puntos más adecuadas (adelante para $x=0.0$, central para $x=0.2$, atrás para $x=0.4$).
2. Sabiendo que $f'''(x) = e^x$, encuentre una cota $M_3$ para $|f'''(x)|$ en el intervalo $[0,0.4]$.
3. Use $M_3$ para calcular las cotas de error teóricas para sus aproximaciones. La derivada real es $f'(x) = e^x - 4x + 3$. Calcule los errores reales y compárelos con las cotas.




## Parte I

#### Diferencias tres puntos:

* Diferencia adelante  para $ f'(x_0) $:


$$
f'(x_0) \approx \frac{-3f(x_0) + 4f(x_0+h) - f(x_0+2h)}{2h}
$$

* Diferencia central para $ f'(x_1) $:


$$
f'(x_1) \approx \frac{f(x_1+h) - f(x_1-h)}{2h}
$$

* Diferencia atrás para $ f'(x_2) $:


$$
f'(x_2) \approx \frac{3f(x_2) - 4f(x_2-h) + f(x_2-2h)}{2h}
$$



```python
import numpy as np

# Datos del problema
x = np.array([0.0, 0.2, 0.4])
fx = np.array([0.00000, 0.74140, 1.3718])
h = 0.2  # Espaciado uniforme

# Fórmulas de 3 puntos
adelante_3pts = (-3*fx[0] + 4*fx[1] - fx[2]) / (2*h)
central_3pts = (fx[2] - fx[0]) / (2*h)
atras_3pts = (3*fx[2] - 4*fx[1] + fx[0]) / (2*h)

print("f'(0.0) ≈",adelante_3pts)
print("f'(0.2) ≈",central_3pts)
print("f'(0.4) ≈",atras_3pts)
```

## Parte II

Dado $ f'''(x) = e^x $, la cota en $[a, b] = [0.0, 0.4]$ es:
$$
M_3 = \max_{x \in [a,b]} |f'''(x)| = e^{b} = e^{0.4} \approx 1.4918
$$

```python
M3 = np.exp(0.4)
print("Cota M_3 para |f'''(x)| en [0, 0.4]:",M3 )
```

## Parte III

* Error para diferencia adelante/atrás:
$$
\text{Cota} = \frac{M_3 h^2}{3}
$$

* Error para diferencia central:
$$
\text{Cota} = \frac{M_3 h^2}{6}
$$

* Comparación con error real:
$$
\text{Error real} = |f'_{\text{apr}}(x) - f'(x)|
$$

```python
# Derivada real
f_prima_real = lambda x: np.exp(x) - 4*x + 3

# Errores reales
error_real_adelante = abs(adelante_3pts - f_prima_real(0.0))
error_real_central = abs(central_3pts - f_prima_real(0.2))
error_real_atras = abs(atras_3pts - f_prima_real(0.4))

# Cotas teóricas
cota_adelante = (M3 * h**2) / 3
cota_central = (M3 * h**2) / 6
cota_atras = (M3 * h**2) / 3

print("Error real f'(0.0):", error_real_adelante,  "| Cota:", cota_adelante)
print("Error real f'(0.2):", error_real_central,  "| Cota:", cota_central)
print("Error real f'(0.4):", error_real_atras, "| Cota:", cota_atras )
```

Las aproximaciones numéricas mostraron mayor precisión en la diferencia central con un error real de 0.0105, mientras que las diferencias adelante y atrás presentaron errores ligeramente mayores
de 0.0295 y 0.0085 respectivamente. La cota $M_3 = e^{0.4} \approx 1.4918$ acotó correctamente los errores.

## **Problema 3**.
1. Derive la fórmula de diferencias hacia adelante de tres puntos para $f'(x_0)$:
$f'(x_0) = \frac{1}{2h}[-3f(x_0) + 4f(x_0+h) - f(x_0+2h)] + E_T(h)$.
Encuentre explícitamente el término de error de truncamiento $E_T(h)$, mostrando que es $O(h^2)$.

## Parte I

* Fórmula de diferencias hacia adelante:
$$
f'(x_0) \approx \frac{-3f(x_0) + 4f(x_0 + h) - f(x_0 + 2h)}{2h}
$$

* Término de error $E_T(h)$:
$$
E_T(h) = \frac{h^2}{3} f'''(\xi), \quad \xi \in [x_0, x_0 + 2h], \quad \text{Error} = O(h^2)
$$

```python
from sympy import symbols, exp, diff, simplify, Function

# Definir variables simbólicas
x0, h = symbols('x0 h')
f = symbols('f', cls=Function)

# Expansiones de Taylor
f_x0 = f(x0)
f_x0_h = f(x0) + h*diff(f(x0), x0) + (h**2/2)*diff(f(x0), x0, 2) + (h**3/6)*diff(f(x0), x0, 3)
f_x0_2h = f(x0) + 2*h*diff(f(x0), x0) + (2*h)**2/2*diff(f(x0), x0, 2) + (2*h)**3/6*diff(f(x0), x0, 3)

# Fórmula de 3 puntos
formula = (-3*f_x0 + 4*f_x0_h - f_x0_2h) / (2*h)
error_term = simplify(formula - diff(f(x0), x0))  # Término de error dominante

print("Fórmula de diferencias hacia adelante (3 puntos):")
print(formula)
print("\nTérmino de error dominante:")
print(error_term)
```

## Parte II


```python
def f(x):
    return x * np.exp(x)

# Puntos y paso
x0 = 0.2
h = 0.1
points = [x0, x0 + h, x0 + 2*h]
fx = [f(p) for p in points]

# Aproximación de la derivada
df_approx = (-3*fx[0] + 4*fx[1] - fx[2]) / (2*h)

print("Aproximación de f'(0.2):", df_approx)
print("Valores de f(x):", fx)
```

## Parte III

* Derivada exacta:
$$
f'(x) = e^x (1 + x) \implies f'(0.2) \approx 1.46568
$$

* Error real:
$$
\text{Error} = |1.46515 - 1.46568| \approx 0.00053
$$

* Cota teórica del error:
$$
|E_T(h)| \leq \frac{h^2}{3} \max_{\xi \in [0.2, 0.4]} |f'''(\xi)|, \quad f'''(x) = e^x (3 + x)
$$

* Cálculo de $M_3$:
$$
M_3 = \max_{x \in [0.2, 0.4]} e^x (3 + x) = e^{0.4} \times 3.4 \approx 5.0736
$$

* Cota final:
$$
|E_T(h)| \leq \frac{0.1^2}{3} \times 5.0736 \approx 0.01691
$$

```python
# Derivada exacta
def f_prime(x):
    return np.exp(x) * (1 + x)

# Error real
error_real = abs(df_approx - f_prime(x0))

# Cálculo de M3 (máxima tercera derivada en [0.2, 0.4])
def f_triple_prime(x):
    return np.exp(x) * (3 + x)

x_vals = np.linspace(0.2, 0.4, 100)
M3 = max(f_triple_prime(x) for x in x_vals)

# Cota teórica
error_bound = (h**2 / 3) * M3

print("\nError real:", error_real)
print("Cota teórica del error:", error_bound)
print("M3:", M3)
```

## **Problema 4**

 Aproxime $\int_1^{1.5} x^2 \ln x \,dx$ usando Python con:

1. Regla del Trapecio simple.
2. Regla de Simpson 1/3 simple.
3. Regla del Punto Medio simple.

El valor de referencia para la integral es aproximadamente $0.19225935773279605$. Reporte los valores de sus aproximaciones, el código Python utilizado y los errores absolutos de cada método. Comente brevemente cuál método fue más preciso y por qué podría esperarse esto.

## Regla del Trapecio Simple

```python
def f(x):
    return x**2 * np.log(x)

# Límites de integración
a, b = 1, 1.5
valor_real = 0.19225935773279605

# Regla del Trapecio
trapecio = (b - a)/2 * (f(a) + f(b))
error_trapecio = abs(trapecio - valor_real)

print("Trapecio:" ,trapecio, "Error:" ,error_trapecio)
```

## Regla de Simpson 1/3 Simple

```python
# Regla de Simpson 1/3
simpson = (b - a)/6 * (f(a) + 4*f((a + b)/2) + f(b))
error_simpson = abs(simpson - valor_real)

print("Simpson:", simpson, "Error:", error_simpson)
```

## Regla del Punto Medio Simple

```python
# Regla del Punto Medio
punto_medio = (b - a) * f((a + b)/2)
error_pmedio = abs(punto_medio - valor_real)

print("Punto Medio:", punto_medio, "Error", error_pmedio)
```

**Conclusión:** La Regla de Simpson 1/3 fue la más precisa debido a su capacidad para capturar mejor el comportamiento no lineal de $x^2lnx$ al utilizar un polinomio de interpolación cuadrático.

## **Problema 5**
 Determine los valores de $n$ (número de subintervalos) y $h$ (ancho del subintervalo) requeridos para aproximar la integral $\int_0^2 e^{2x}\sin(3x)dx$ dentro de una tolerancia de $10^{-4}$ utilizando:

1. La regla compuesta del Trapecio. El término de error es $E_T = -\frac{(b-a)h^2}{12}f''(\xi)$.
2. La regla compuesta de Simpson 1/3. El término de error es $E_S = -\frac{(b-a)h^4}{180}f^{(4)}(\xi)$.

Para esto, necesitará calcular $f''(x)$ y $f^{(4)}(x)$ y estimar cotas para sus valores absolutos en el intervalo $[0,2]$.

## Parte I

```python
import sympy as sp

# Definir la variable y función
x = sp.Symbol('x')
f = sp.exp(2*x) * sp.sin(3*x)

# Derivadas simbólicas
f2 = sp.diff(f, x, 2)
f4 = sp.diff(f, x, 4)

# Convertir a funciones evaluables
f2_func = sp.lambdify(x, f2, modules=['numpy'])
f4_func = sp.lambdify(x, f4, modules=['numpy'])

# Encontrar máximos absolutos en el intervalo [0,2]
import numpy as np
x_vals = np.linspace(0, 2, 1000)
max_f2 = np.max(np.abs(f2_func(x_vals)))
max_f4 = np.max(np.abs(f4_func(x_vals)))

print("Máximo de |f''(x)| en [0,2]:", max_f2)
print("Máximo de |f⁽⁴⁾(x)| en [0,2]", max_f4)
```

## Parte II

```python
# Tolerancia requerida
tolerance = 1e-4
a, b = 0, 2

# Trapecio: Error_T = (b-a) * h^2 / 12 * max|f''|
h_trap = np.sqrt(12 * tolerance / ((b - a) * max_f2))
n_trap = int(np.ceil((b - a) / h_trap))
h_trap = (b - a) / n_trap

# Simpson: Error_S = (b-a) * h^4 / 180 * max|f⁽⁴⁾|
h_simp = (180 * tolerance / ((b - a) * max_f4))**(1/4)
n_simp = int(np.ceil((b - a) / h_simp))
if n_simp % 2 != 0:
    n_simp += 1  # Simpson requiere n par
h_simp = (b - a) / n_simp

print(f"[Trapecio] n = {n_trap}, h = {h_trap:.6f}")
print(f"[Simpson] n = {n_simp}, h = {h_simp:.6f}")
```

## **Problema 6**

 Considere la integral $I = \int_0^{8} \sqrt{1+\cos^2 x}dx$.

1. Calcule $R_{1,1}, R_{2,1}, R_{3,1}, R_{4,1}$ para la tabla de integración de Romberg. (Estos son los valores de la Regla del Trapecio Compuesta con $n=1, 2, 4, 8$ subintervalos respectivamente).
2. Use estos valores para construir la tabla de Romberg hasta $R_{4,4}$.
3. Explique por qué esta integral, a pesar de tener un integrando suave y periódico, podría presentar desafíos para la convergencia rápida de Romberg en un intervalo como $[0,8]$ (que es más de un período de $\cos x$). ¿Qué indica el comportamiento de los valores en la tabla de Romberg sobre la efectividad del método en este caso?

```python
# Definimos la función a integrar
f = lambda x: np.sqrt(1 + np.cos(x)**2)
a, b = 0, 8

# Inicializamos la tabla de Romberg
max_level = 4
R = np.zeros((max_level + 1, max_level + 1))

# Paso 1: Calcular R[k, 1] = R_{k,1} usando la regla del trapecio compuesta
for k in range(1, max_level + 1):
    n = 2**(k - 1)
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    R[k, 1] = h * (0.5 * y[0] + np.sum(y[1:-1]) + 0.5 * y[-1])

# Paso 2: Usar extrapolación de Richardson para llenar el resto de la tabla
for j in range(2, max_level + 1):
    for k in range(j, max_level + 1):
        R[k, j] = R[k, j - 1] + (R[k, j - 1] - R[k - 1, j - 1]) / (4**(j - 1) - 1)

# Imprimir tabla formateada
from tabulate import tabulate

headers = ["k \\ j"] + [str(j) for j in range(1, max_level + 1)]
table_data = []
for k in range(1, max_level + 1):
    row = [str(k)]
    for j in range(1, max_level + 1):
        if j <= k:
            row.append(f"{R[k, j]:.8f}")
        else:
            row.append("")
    table_data.append(row)

print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

**Conclusiones:**

 Romberg no necesariamente converge rápidamente en intervalos largos que abarcan múltiples periodos. Esto ocurre porque los errores de la integración compuesta pueden acumularse constructivamente en regiones con oscilaciones periódicas.

En el intervalo $[0,8]$, que incluye más de un período de $\cos x$, los subintervalos iniciales (como $n=1$ o $n=2$) no capturan bien el comportamiento oscilatorio, por lo que los valores de $R_{k,j}$ pueden converger lentamente o incluso de forma inestable al inicio.
